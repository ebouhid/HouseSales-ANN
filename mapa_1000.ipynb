```python
import pandas as pd
import folium
```

```python
dataDf = pd.read_csv(open('/content/drive/MyDrive/Arquivos de UCs/UCs/7 semestre/IA/Praticas/Pratica 2/export.csv'))#abre o arquivo csv com as informacoes das casas
dataDf.head()
```

```python
#Cria o mapa
mediaLat = dataDf['lat'].mean()
mediaLong = dataDf['long'].mean()

mapObj = folium.Map(location = [mediaLat,mediaLong],zoom_start=5,tiles='stamenterrain')

#Localiza a regiÃ£o solicitada
borderStyle={
    "color": "green",
    "weight": 2,
    "fill": False
}

folium.GeoJson(data='/content/drive/MyDrive/Data/House_Sale/us-states.json',
               name = 'Washington',
               style_function=lambda x: borderStyle).add_to(mapObj)

plantsLayer = folium.FeatureGroup("Power Plants")

dataDf.sample(frac=1)#Embaralha o dataframe, para trazer aleatoriedade
itrs = int(1000)

#Plota as propriedades no mapa e mostra seus atributos quando passar o mouse acima
#No plot dos atributos, existe uma consistÃªncia e padronizaÃ§Ã£o: atributos semelhantes estÃ£o prÃ³ximos entre si.
#Por exemplo, price e new_price estÃ£o prÃ³ximos, as Ã¡reas tambÃ©m e assim por diante
for itr in range(itrs):
  latStr = dataDf.iloc[itr]['lat']
  longStr = dataDf.iloc[itr]['long']
  priceStr = dataDf.iloc[itr]['price']
  idStr = dataDf.iloc[itr]['id']
  floorsStr = dataDf.iloc[itr]['floors']
  bathroomsStr = dataDf.iloc[itr]['bathrooms']
  livingAreaStr = dataDf.iloc[itr]['sqft_living']
  lotAreaStr = dataDf.iloc[itr]['sqft_lot']
  livingArea15Str = dataDf.iloc[itr]['sqft_living15']
  lotArea15Str = dataDf.iloc[itr]['sqft_lot15']
  waterFrontStr = dataDf.iloc[itr]['waterfront']
  conditionStr = dataDf.iloc[itr]['condition']
  color = "blue" if latStr>=mediaLat else "red"
  newPriceStr = dataDf.iloc[itr]['new_price']
  bedroomsStr = dataDf.iloc[itr]['bedrooms']
  gradeStr = dataDf.iloc[itr]['grade']
  aboveStr = dataDf.iloc[itr]['sqft_above']
  basementStr = dataDf.iloc[itr]['sqft_basement']
  yearBuiltStr = dataDf.iloc[itr]['yr_built']
  yearRenovatedStr = dataDf.iloc[itr]['yr_renovated']
  zipcodeStr = dataDf.iloc[itr]['zipcode']
  folium.Marker(
      location = [latStr,longStr],
      color = color,
      weight = 2,
      fill = True,
      fill_opacity = 0.1,
      tooltip=f"""<b>ID:</b> {idStr}<br>
      <b>Price:</b> US${priceStr}<br>
      <b>New price:</b> US${newPriceStr:.2f}<br>
      <b>Floors:</b> {floorsStr}<br>
      <b>Bedrooms:</b> {bedroomsStr}<br>
      <b>Bathrooms:</b> {bathroomsStr}<br>
      <b>Waterfront:</b> {waterFrontStr}<br>
      <b>Living Area:</b> {livingAreaStr}<br>
      <b>Lot Area:</b> {lotAreaStr}<br>
      <b>Living Area 15:</b> {livingArea15Str}<br>
      <b>Lot Area 15:</b> {lotArea15Str}<br>
      <b>Above:</b> {aboveStr}<br>
      <b>Basement:</b> {basementStr}<br>
      <b>Year Built:</b> {yearBuiltStr}<br>
      <b>Year Renovated:</b> {yearRenovatedStr}<br>
      <b>Condition:</b> {conditionStr}<br>
      <b>Grade:</b> {gradeStr}<br>
      <b>Latitude:</b> {latStr}<br>
      <b>Longitude:</b> {longStr}<br>
      <b>Zipcode:</b> {zipcodeStr}<br>
      """
  ).add_to(mapObj)

#add a tile layer
folium.TileLayer('OpenStreetMap',
                 attr='OpenStreetMap').add_to(mapObj)
#add tile layer (from external provider)
folium.TileLayer('https://{s}.tile.opentopomap.org/{z}/{x}/{y}.png',
                 name = 'TopoMap',
                 attr = 'OpenTopoMap').add_to(mapObj)

folium.LayerControl().add_to(mapObj)

mapObj.save('mapa_1000.html')

mapObj
```
